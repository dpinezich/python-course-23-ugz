## Styling & Themes
Wir verwenden Visualisierungen, um Erkenntnisse aus einer riesigen Datenmenge zu gewinnen oder unsere Erkenntnisse Interessengruppen zu prÃ¤sentieren, und eine ansprechende Gestaltung dieser Visualisierungen kann sicherlich bei beidem helfen.

Menschen konzentrieren sich eher auf schÃ¶ne und attraktive Visualisierungen als auf langweilige Diagramme, daher kann das Styling als eine wichtige Komponente der Datenvisualisierung angesehen werden.

Die Matplotlib-Bibliothek ist in hohem Masse anpassbar, aber es kann schwierig sein, die richtigen Einstellungen zu finden. Hier bietet Seaborn eine Highlevel-Schnittstelle an.

## Example

```python
from matplotlib import pyplot as plt_norm
plt_norm.scatter(df.speeding,df.alcohol)
plt_norm.show()
```

## Ein anderes Beispiel

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.scatter(df.speeding,df.alcohol)
sns.set_style("darkgrid")
plt.show()
```

Die beiden obigen Abbildungen zeigen den Unterschied zwischen den Standarddiagrammen von Matplotlib und Seaborn. Die Darstellung der Daten ist die Gleiche, aber es gibt einen kleinen Unterschied in der Gestaltung der Diagramme.

Seaborn unterstÃ¼tzt verschiedene Themen, die das Styling der Plots sehr einfach machen und viel Zeit sparen kÃ¶nnen. Mit der set_style()-Funktion von Seaborn kÃ¶nnen wir jedes der in der Seaborn-Bibliothek verfÃ¼gbaren Themen einstellen. Hier sind ein paar der beliebtesten Themen:
* Darkgrid
* Whitegrid
* Dark
* White
* Ticks

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.scatter(df.speeding,df.alcohol)
sns.set_style("dark")
plt.show()
```

Seaborn ermÃ¶glicht es uns auch, einzelne Elemente unserer Diagramme zu steuern,  so kÃ¶nnen wir die Skalierung dieser Elemente oder der Darstellung mit der Funktion set_context() beeinflussen. #### Barplot+

```python
import matplotlib.pyplot as plt
import seaborn as sns
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic,
            palette = 'PuRd',
            order = ['male', 'female'],
            capsize = 0.05,
            saturation = 8,
            errcolor = 'gray', errwidth = 2,
            ci = 'sd'
            )
plt.legend()
plt.show()
```

#### Countplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('paper')

# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.countplot(x = 'class', hue = 'who', data = titanic, palette = 'magma')
plt.title('Survivors')
plt.show()
```

#### Pointplot

```python
# importing required packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("tips")
sns.pointplot(x="day", y="tip", data=data)
plt.show()
```
